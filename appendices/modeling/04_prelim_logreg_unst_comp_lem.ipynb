```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
import tensorflow as tf
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/down_sample.csv')[['Subreddit', 'unst_comp_lem']]
```

```python
df.head()
```

```python
X = df.drop(columns='Subreddit')
y = df['Subreddit']

X_train, X_test, y_train, y_test=train_test_split(X, y, stratify = y, random_state=42)
```

```python
cvec = CountVectorizer()
cvec.fit(X_train['unst_comp_lem'])
X_train_cv = pd.DataFrame(cvec.transform(X_train['unst_comp_lem']).toarray(), columns = cvec.get_feature_names())
X_test_cv = pd.DataFrame(cvec.transform(X_test['unst_comp_lem']).toarray(), columns = cvec.get_feature_names())
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train_cv, y_train)
```

```python
lr.score(X_train_cv, y_train)
```

```python
cross_val_score(lr, X_train_cv, y_train, cv=5).mean()
```

```python
lr.score(X_test_cv, y_test)
```

```python
tvec = TfidfVectorizer()
tvec.fit(X_train['unst_comp_lem'])
X_train_tv = pd.DataFrame(tvec.transform(X_train['unst_comp_lem']).todense(), columns = tvec.get_feature_names())
X_test_tv = pd.DataFrame(tvec.transform(X_test['unst_comp_lem']).todense(), columns = tvec.get_feature_names())
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train_tv, y_train)
```

```python
lr.score(X_train_tv, y_train)
```

```python
cross_val_score(lr, X_train_tv, y_train, cv=5).mean()
```

```python
lr.score(X_test_tv, y_test)
```
